# What are the most in-demand job roles?

```python
sns.countplot(x ='Job Title', data=df, hue="Job Title", palette="plasma")
plt.xticks(rotation = 70)
plt.xlabel("Job Title", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.title("Most in-demand job roles", fontsize=17, color="red")
plt.show()
```

**Q**) Which companies are posting the most job openings?

```python
sns.countplot(x ='Company Name', data=df, hue="Company Name", palette="plasma")
plt.xticks(rotation = 70)
plt.xlabel("Company Name", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.title("Company Name", fontsize=17, color="red")
plt.show()
```

What is the distribution of job types?
**bold text**

```python
gb = df.groupby('Job Type').agg({"Job ID":"count"})
plt.pie(gb["Job ID"], labels=gb.index, autopct='%1.1f%%')
plt.title("Distribution of Job Types")
plt.show()
gb.head()
```

# What is the distribution of salary ranges across different job roles?

```python
plt.figure(figsize = (10,4)) #to change the size of the figure
sns.countplot(x = "Job Title", data = df, hue = "Salary Range", palette = "viridis", hue_order = ["3-5LPA", "5-8 LPA", "8-12 LPA", "12-20 LPA", "20+ LPA"])
plt.xticks(rotation = 70)
plt.legend(bbox_to_anchor = (1,1)) #bbox is used to keep the legend outside the graph
plt.show()
```

```python
company_postings = df['Company Name'].value_counts() # Changed job_data to df
plt.figure(figsize=(10, 6))
sns.barplot(x=company_postings.head(10).values, y=company_postings.head(10).index, palette="viridis")
plt.title("Top 10 Companies Posting the Most Job Openings", fontsize=16)
plt.xlabel("Number of Job Postings", fontsize=12)
plt.ylabel("Company Name", fontsize=12)
plt.tight_layout()
plt.show()
```

# Distribution of Salary Ranges Across Different Job Roles

```python
salary_distribution = df.groupby('Job Title')['Salary Range'].value_counts().unstack(fill_value=0)
plt.figure(figsize=(10, 6))
salary_distribution.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Distribution of Salary Ranges Across Different Job Roles', fontsize=16)
plt.xlabel('Job Title', fontsize=12)
plt.xticks(rotation = 70)
plt.ylabel('Number of Postings', fontsize=12)
plt.legend(title='Salary Range', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

# Salary Range with the Highest Number of Job Postings

```python
salary_range_count = df['Salary Range'].value_counts()
salary_range_count.plot(kind='bar', color='lightgreen', figsize=(8, 6))
plt.title('Salary Range with the Highest Number of Job Postings')
plt.xlabel('Salary Range')
plt.ylabel('Number of Postings')
plt.show()
```

# Cities with the Highest Number of Job Opportunities

```python
job_location_count = df['Job Location'].value_counts()
job_location_count.head(10).plot(kind='bar', color='coral', figsize=(10, 6))
plt.title('Cities with the Highest Number of Job Opportunities')
plt.xlabel('City')
plt.ylabel('Number of Postings')
plt.show()
```

# Most Common Job Types in Each Location

```python
location_job_type = df.groupby('Job Location')['Job Type'].value_counts().unstack(fill_value=0)
location_job_type.loc[location_job_type.index[:5]].plot(kind='bar', stacked=True, figsize=(12, 6), colormap='coolwarm')
plt.title('Most Common Job Types in Each Location')
plt.xlabel('City')
plt.ylabel('Number of Postings')
plt.legend(title='Job Type')
plt.show()
```

# Most Frequently Required Skills

```python
skills = df['Skills Required'].str.split(', ').explode()
skills_count = skills.value_counts()
skills_count.head(10).plot(kind='bar', color='orange', figsize=(10, 6))
plt.title('Most Frequently Required Skills')
plt.xlabel('Skills')
plt.ylabel('Frequency')
plt.show()
```

# Correlation Between Education and Salary Ranges

```python
education_salary = pd.crosstab(df['Education Requirement'], df['Salary Range'])
sns.heatmap(education_salary, annot=True, cmap='Blues', fmt='d')
plt.title('Correlation Between Education Levels and Salary Ranges')
plt.xlabel('Salary Range')
plt.ylabel('Education Requirement')
plt.show()
```

# Most Demanded Education Level for Job Roles

```python
education_job_title = df.groupby('Job Title')['Education Requirement'].value_counts()
education_job_title.loc['Software Engineer'].plot(kind='bar', color='purple', figsize=(8, 6))
plt.title('Education Levels for Software Engineer Roles')
plt.xlabel('Education Requirement')
plt.ylabel('Frequency')
plt.show()
```

# Variation of Job Postings Over Time

```python
df['Posted Date'] = pd.to_datetime(df['Posted Date'])
postings_by_month = df['Posted Date'].dt.month.value_counts().sort_index()
postings_by_month.plot(kind='line', marker='o', figsize=(10, 6), color='green')
plt.title('Number of Job Postings Over Time (by Month)')
plt.xlabel('Month')
plt.ylabel('Number of Postings')
plt.show()
```

# Average Duration Between Posting and Deadline

```python
df['Application Deadline'] = pd.to_datetime(df['Application Deadline'])
df['Duration'] = (df['Application Deadline'] - df['Posted Date']).dt.days
df['Duration'].plot(kind='hist', bins=20, color='teal', figsize=(8, 6))
plt.title('Distribution of Duration Between Posting and Deadline')
plt.xlabel('Duration (Days)')
plt.ylabel('Frequency')
plt.show()
```

# Job Portals Attracting the Most Applicants


```python
applicants_by_portal = df.groupby('Job Portal')['Number of Applicants'].sum()
applicants_by_portal.plot(kind='bar', color='pink', figsize=(8, 6))
plt.title('Job Portals Attracting the Most Applicants')
plt.xlabel('Job Portal')
plt.ylabel('Number of Applicants')
plt.show()
```

# Job Titles Receiving the Most Applications

```python
applicants_by_title = df.groupby('Job Title')['Number of Applicants'].sum()
applicants_by_title.sort_values(ascending=False).head(10).plot(kind='bar', color='cyan', figsize=(10, 6))
plt.title('Job Titles Receiving the Most Applications')
plt.xlabel('Job Title')
plt.ylabel('Number of Applicants')
plt.show()
```

# Distribution of Job Postings by Company Size

```python
company_size_count = df['Company Size'].value_counts()
company_size_count.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('muted'), startangle=90, figsize=(8, 6))
plt.title('Distribution of Job Postings by Company Size')
plt.ylabel('')
plt.show()
```

# Companies Offering the Most Remote Work Opportunities

```python
remote_jobs = df[df['Remote/Onsite'] == 'Remote']
remote_companies = remote_jobs['Company Name'].value_counts()
remote_companies.head(10).plot(kind='bar', color='lime', figsize=(10, 6))
plt.title('Top Companies Offering Remote Work Opportunities')
plt.xlabel('Company Name')
plt.ylabel('Number of Remote Jobs')
plt.show()
```
